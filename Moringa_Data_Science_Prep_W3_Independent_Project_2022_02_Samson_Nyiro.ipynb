```python
#Checking for Duplicate Values
telecom1.duplicated()
telecom1.drop_duplicates()

#Checking for Null Values, getting rid of them and looking for missing values
telecom1.isnull()
missing_dt1 = telecom1.dropna()
missing_dt1

# #Grouping of the Data
telecom_group = missing_dt1['CELL_ID'].groupby(missing_dt1['PRODUCT'])


# #Drop Items not useful to the research
missing_dt1.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)

# #Renaming Columns
telecom1.rename(columns = {'PRODUTC':'PRODUCT', 'VALUE':'BILLING',
                              }, inplace = True)
print(missing_dt1.columns)
print(telecom1.columns)
```

```python
#Splitting the Values on the DATETIME column to have time and Date separately
telecom1['DATETIME'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split('-'))]))
```
